```python
import asyncio
import edge_tts
import json
import os
import textwrap
import re
from moviepy.editor import *
from PIL import Image, ImageDraw, ImageFont

# Import ‡∏ï‡∏±‡∏ß‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á PyThaiNLP
from pythainlp.tokenize import word_tokenize
```

```python
# ====================================================
# 1. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥ (‡∏â‡∏ö‡∏±‡∏ö‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç: ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ text, font, width)
# ====================================================
def thai_text_wrap(text, font, max_width):
    """
    ‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥‡πÇ‡∏î‡∏¢‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ (‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°, ‡∏ü‡∏≠‡∏ô‡∏ï‡πå, ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Å‡∏ß‡πâ‡∏≤‡∏á‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î)
    """
    if not text: return []

    # ‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏î‡∏¥‡∏°‡∏Å‡πà‡∏≠‡∏ô
    paragraphs = text.split('\n')
    final_lines = []

    for para in paragraphs:
        if not para:
            final_lines.append("")
            continue

        words = word_tokenize(para, engine='newmm')
        current_line = ""

        for word in words:
            test_line = current_line + word

            # ‡∏ß‡∏±‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Å‡∏ß‡πâ‡∏≤‡∏á (‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö‡∏ó‡∏±‡πâ‡∏á Pillow ‡πÄ‡∏Å‡πà‡∏≤‡πÅ‡∏•‡∏∞‡πÉ‡∏´‡∏°‡πà)
            if hasattr(font, 'getlength'):
                w = int(font.getlength(test_line))
            else:
                w = font.getsize(test_line)[0]

            if w <= max_width:
                current_line = test_line
            else:
                if current_line: final_lines.append(current_line)

                # ‡∏Å‡∏£‡∏ì‡∏µ‡∏Ñ‡∏≥‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏¢‡∏≤‡∏ß‡∏Å‡∏ß‡πà‡∏≤‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î
                if hasattr(font, 'getlength'):
                    word_w = int(font.getlength(word))
                else:
                    word_w = font.getsize(word)[0]

                if word_w > max_width:
                    final_lines.append(word)
                    current_line = ""
                else:
                    current_line = word

        if current_line: final_lines.append(current_line)

    return final_lines
```

```python
# ====================================================
# 2. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ß‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏° (‡∏â‡∏ö‡∏±‡∏ö‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç: ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏ó‡∏±‡πâ‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÅ‡∏•‡∏∞‡∏ü‡∏≠‡∏ô‡∏ï‡πå)
# ====================================================
def draw_text_box(draw, text, x, y, w, h, font_or_size=None, min_font_size=12, fill=(0,0,0), align='left', valign='center', **kwargs):
    """
    ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ß‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏≠‡∏±‡∏à‡∏â‡∏£‡∏¥‡∏¢‡∏∞ (Hybrid):
    - ‡∏ñ‡πâ‡∏≤‡∏™‡πà‡∏á Font Object (‡∏à‡∏≤‡∏Å Question Slide) -> ‡∏ß‡∏≤‡∏î‡πÄ‡∏•‡∏¢
    - ‡∏ñ‡πâ‡∏≤‡∏™‡πà‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (‡∏à‡∏≤‡∏Å Answer Slide) -> Auto-Fit ‡∏•‡∏î‡∏Ç‡∏ô‡∏≤‡∏î‡πÉ‡∏´‡πâ‡∏û‡∏≠‡∏î‡∏µ
    """
    # --- ‡∏Å‡∏£‡∏ì‡∏µ 1: ‡πÅ‡∏ö‡∏ö‡πÄ‡∏Å‡πà‡∏≤ (Question Slide ‡∏™‡πà‡∏á Font Object ‡∏°‡∏≤) ---
    if hasattr(font_or_size, 'getsize') or hasattr(font_or_size, 'getlength'):
        font = font_or_size
        # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å thai_text_wrap ‡πÅ‡∏ö‡∏ö‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á (text, font, width)
        lines = thai_text_wrap(text, font, w - 20)

        # ‡πÉ‡∏ä‡πâ‡∏£‡∏∞‡∏¢‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î 1.5 ‡πÄ‡∏ó‡πà‡∏≤ (‡∏™‡∏π‡∏ï‡∏£‡∏Å‡∏•‡∏≤‡∏á‡πÜ)
        line_height = int(font.size * 1.5)
        total_text_height = len(lines) * line_height

        if valign == 'center': start_y = y + (h - total_text_height) / 2
        elif valign == 'top': start_y = y + 15
        else: start_y = y + 15

        if start_y < y: start_y = y + 5

        current_y = start_y
        for line in lines:
            if current_y + line_height > y + h + 5: break

            draw_x = x + 10
            if align == 'center':
                if hasattr(font, 'getlength'): lw = int(font.getlength(line))
                else: lw = font.getsize(line)[0]
                draw_x = x + (w - lw) / 2

            draw.text((draw_x, current_y), line, fill=fill, font=font)
            current_y += line_height
        return

    # --- ‡∏Å‡∏£‡∏ì‡∏µ 2: ‡πÅ‡∏ö‡∏ö‡πÉ‡∏´‡∏°‡πà (Answer Slide ‡∏™‡πà‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç/None ‡∏°‡∏≤) ---
    initial_font_size = font_or_size if isinstance(font_or_size, int) else 40

    best_font = None
    best_lines = []
    best_line_height = 0
    best_total_h = 0

    # Loop ‡∏´‡∏≤‡∏Ç‡∏ô‡∏≤‡∏î‡∏ü‡∏≠‡∏ô‡∏ï‡πå (Auto-Fit)
    for size in range(initial_font_size, min_font_size - 1, -2):
        font = get_font(size, bold=False)
        # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å thai_text_wrap ‡πÅ‡∏ö‡∏ö‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
        lines = thai_text_wrap(text, font, w - 20)

        # ‡πÉ‡∏ä‡πâ‡∏£‡∏∞‡∏¢‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î 1.8 ‡πÄ‡∏ó‡πà‡∏≤ (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏â‡∏•‡∏¢ ‡∏Å‡∏±‡∏ô‡∏ó‡∏±‡∏ö‡πÅ‡∏ô‡πà‡∏ô‡∏≠‡∏ô)
        line_height = int(size * 1.8)
        total_h = len(lines) * line_height

        if total_h <= h:
            best_font = font
            best_lines = lines
            best_line_height = line_height
            best_total_h = total_h
            break

    if best_font is None: # Fallback
        best_font = get_font(min_font_size, bold=False)
        best_lines = thai_text_wrap(text, best_font, w - 20)
        best_line_height = int(min_font_size * 1.8)
        best_total_h = len(best_lines) * best_line_height

    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Y
    if valign == 'center': start_y = y + (h - best_total_h) / 2
    elif valign == 'top': start_y = y + 15
    else: start_y = y + h - best_total_h - 15

    if start_y < y: start_y = y + 5

    # ‡∏ß‡∏≤‡∏î
    current_y = start_y
    for line in best_lines:
        if current_y + best_line_height > y + h + 5: break

        draw_x = x + 10
        if align == 'center':
            if hasattr(best_font, 'getlength'): lw = int(best_font.getlength(line))
            else: lw = best_font.getsize(line)[0]
            draw_x = x + (w - lw) / 2

        draw.text((draw_x, current_y), line, fill=fill, font=best_font)
        current_y += best_line_height
```

```python
def get_font(size, bold=False):
    font_file = FONT_BOLD if bold else FONT_REG
    try:
        return ImageFont.truetype(font_file, size)
    except:
        return ImageFont.load_default()
```

```python
# ====================================================
# 1. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥ (‡∏â‡∏ö‡∏±‡∏ö‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç: ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ text, font, width)
# ====================================================
def thai_text_wrap(text, font, max_width):
    """
    ‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥‡πÇ‡∏î‡∏¢‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ (‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°, ‡∏ü‡∏≠‡∏ô‡∏ï‡πå, ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Å‡∏ß‡πâ‡∏≤‡∏á‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î)
    """
    if not text: return []

    # ‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏î‡∏¥‡∏°‡∏Å‡πà‡∏≠‡∏ô
    paragraphs = text.split('\n')
    final_lines = []

    for para in paragraphs:
        if not para:
            final_lines.append("")
            continue

        words = word_tokenize(para, engine='newmm')
        current_line = ""

        for word in words:
            test_line = current_line + word

            # ‡∏ß‡∏±‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Å‡∏ß‡πâ‡∏≤‡∏á (‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö‡∏ó‡∏±‡πâ‡∏á Pillow ‡πÄ‡∏Å‡πà‡∏≤‡πÅ‡∏•‡∏∞‡πÉ‡∏´‡∏°‡πà)
            if hasattr(font, 'getlength'):
                w = int(font.getlength(test_line))
            else:
                w = font.getsize(test_line)[0]

            if w <= max_width:
                current_line = test_line
            else:
                if current_line: final_lines.append(current_line)

                # ‡∏Å‡∏£‡∏ì‡∏µ‡∏Ñ‡∏≥‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏¢‡∏≤‡∏ß‡∏Å‡∏ß‡πà‡∏≤‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î
                if hasattr(font, 'getlength'):
                    word_w = int(font.getlength(word))
                else:
                    word_w = font.getsize(word)[0]

                if word_w > max_width:
                    final_lines.append(word)
                    current_line = ""
                else:
                    current_line = word

        if current_line: final_lines.append(current_line)

    return final_lines
```

```python
# ====================================================
# 2. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ß‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏° (‡∏â‡∏ö‡∏±‡∏ö‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç: ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏ó‡∏±‡πâ‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÅ‡∏•‡∏∞‡∏ü‡∏≠‡∏ô‡∏ï‡πå)
# ====================================================
def draw_text_box(draw, text, x, y, w, h, font_or_size=None, min_font_size=12, fill=(0,0,0), align='left', valign='center', **kwargs):
    """
    ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ß‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏≠‡∏±‡∏à‡∏â‡∏£‡∏¥‡∏¢‡∏∞ (Hybrid):
    - ‡∏ñ‡πâ‡∏≤‡∏™‡πà‡∏á Font Object (‡∏à‡∏≤‡∏Å Question Slide) -> ‡∏ß‡∏≤‡∏î‡πÄ‡∏•‡∏¢
    - ‡∏ñ‡πâ‡∏≤‡∏™‡πà‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (‡∏à‡∏≤‡∏Å Answer Slide) -> Auto-Fit ‡∏•‡∏î‡∏Ç‡∏ô‡∏≤‡∏î‡πÉ‡∏´‡πâ‡∏û‡∏≠‡∏î‡∏µ
    """
    # --- ‡∏Å‡∏£‡∏ì‡∏µ 1: ‡πÅ‡∏ö‡∏ö‡πÄ‡∏Å‡πà‡∏≤ (Question Slide ‡∏™‡πà‡∏á Font Object ‡∏°‡∏≤) ---
    if hasattr(font_or_size, 'getsize') or hasattr(font_or_size, 'getlength'):
        font = font_or_size
        # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å thai_text_wrap ‡πÅ‡∏ö‡∏ö‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á (text, font, width)
        lines = thai_text_wrap(text, font, w - 20)

        # ‡πÉ‡∏ä‡πâ‡∏£‡∏∞‡∏¢‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î 1.5 ‡πÄ‡∏ó‡πà‡∏≤ (‡∏™‡∏π‡∏ï‡∏£‡∏Å‡∏•‡∏≤‡∏á‡πÜ)
        line_height = int(font.size * 1.5)
        total_text_height = len(lines) * line_height

        if valign == 'center': start_y = y + (h - total_text_height) / 2
        elif valign == 'top': start_y = y + 15
        else: start_y = y + 15

        if start_y < y: start_y = y + 5

        current_y = start_y
        for line in lines:
            if current_y + line_height > y + h + 5: break

            draw_x = x + 10
            if align == 'center':
                if hasattr(font, 'getlength'): lw = int(font.getlength(line))
                else: lw = font.getsize(line)[0]
                draw_x = x + (w - lw) / 2

            draw.text((draw_x, current_y), line, fill=fill, font=font)
            current_y += line_height
        return

    # --- ‡∏Å‡∏£‡∏ì‡∏µ 2: ‡πÅ‡∏ö‡∏ö‡πÉ‡∏´‡∏°‡πà (Answer Slide ‡∏™‡πà‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç/None ‡∏°‡∏≤) ---
    initial_font_size = font_or_size if isinstance(font_or_size, int) else 40

    best_font = None
    best_lines = []
    best_line_height = 0
    best_total_h = 0

    # Loop ‡∏´‡∏≤‡∏Ç‡∏ô‡∏≤‡∏î‡∏ü‡∏≠‡∏ô‡∏ï‡πå (Auto-Fit)
    for size in range(initial_font_size, min_font_size - 1, -2):
        font = get_font(size, bold=False)
        # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å thai_text_wrap ‡πÅ‡∏ö‡∏ö‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
        lines = thai_text_wrap(text, font, w - 20)

        # ‡πÉ‡∏ä‡πâ‡∏£‡∏∞‡∏¢‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î 1.8 ‡πÄ‡∏ó‡πà‡∏≤ (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏â‡∏•‡∏¢ ‡∏Å‡∏±‡∏ô‡∏ó‡∏±‡∏ö‡πÅ‡∏ô‡πà‡∏ô‡∏≠‡∏ô)
        line_height = int(size * 1.8)
        total_h = len(lines) * line_height

        if total_h <= h:
            best_font = font
            best_lines = lines
            best_line_height = line_height
            best_total_h = total_h
            break

    if best_font is None: # Fallback
        best_font = get_font(min_font_size, bold=False)
        best_lines = thai_text_wrap(text, best_font, w - 20)
        best_line_height = int(min_font_size * 1.8)
        best_total_h = len(best_lines) * best_line_height

    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Y
    if valign == 'center': start_y = y + (h - best_total_h) / 2
    elif valign == 'top': start_y = y + 15
    else: start_y = y + h - best_total_h - 15

    if start_y < y: start_y = y + 5

    # ‡∏ß‡∏≤‡∏î
    current_y = start_y
    for line in best_lines:
        if current_y + best_line_height > y + h + 5: break

        draw_x = x + 10
        if align == 'center':
            if hasattr(best_font, 'getlength'): lw = int(best_font.getlength(line))
            else: lw = best_font.getsize(line)[0]
            draw_x = x + (w - lw) / 2

        draw.text((draw_x, current_y), line, fill=fill, font=best_font)
        current_y += best_line_height
```

```python
# ==========================================
# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏†‡∏≤‡∏û Slide (Layout ‡πÅ‡∏ö‡∏ö Fixed Pixel 1920x1080)
# ==========================================
def create_question_slide(q_id, narrator_text, choices_eng, filename):
    W, H = 1920, 1080
    img = Image.new('RGB', (W, H), color=(255, 255, 255))
    draw = ImageDraw.Draw(img)

    # --- Title: Question X ---
    # ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á: ‡∏ã‡πâ‡∏≤‡∏¢‡∏ö‡∏ô (50, 50) ‡∏Ç‡∏ô‡∏≤‡∏î 80px
    draw.text((80, 60), f"Question {q_id}", fill=(0,0,0), font=get_font(100, bold=True))

    # --- Narrator Text ---
    # ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á: (80, 250) ‡∏Ç‡∏ô‡∏≤‡∏î 70px
    draw_text_box(draw, narrator_text, 80, 220, 1760, 300, get_font(70, bold=False))

    # --- Line ---
    draw.line((80, 450, 1840, 450), fill=(200,200,200), width=4)

    # --- Choices ---
    # ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ó‡∏µ‡πà Y=500, ‡∏´‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô 130px
    y_start = 500
    for choice in choices_eng:
        draw.text((120, y_start), choice, fill=(0,0,0), font=get_font(60, bold=False))
        y_start += 130

    img.save(filename)
    return filename
```

```python
def create_detailed_answer_slide(q_data, filename):
    W, H = 1920, 1080
    img = Image.new('RGB', (W, H), color=(255, 255, 255))
    draw = ImageDraw.Draw(img)

    # --- Title ---
    draw.text((50, 30), f"Answer {q_data['id']}", fill=(0,0,0), font=get_font(90, bold=True))

    # --- Table Config (Fixed Pixels) ---
    table_x = 50
    table_y = 160
    table_w = 1820 # ‡πÄ‡∏ï‡πá‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Å‡∏ß‡πâ‡∏≤‡∏á‡∏•‡∏ö margin

    # ‡πÅ‡∏ö‡πà‡∏á‡∏™‡∏±‡∏î‡∏™‡πà‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå: Script(30%) Thai(30%) Choices(20%) Answer(20%)
    col_w = [int(table_w*0.3), int(table_w*0.3), int(table_w*0.2), int(table_w*0.2)]

    header_h = 80
    row_h = 200 # ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏π‡∏á‡∏ï‡πà‡∏≠‡πÅ‡∏ñ‡∏ß

    font_head = get_font(40, bold=True)
    font_content = get_font(36, bold=False) # ‡∏Ç‡∏ô‡∏≤‡∏î‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡∏±‡∏á‡∏™‡∏∑‡∏≠‡πÉ‡∏ô‡∏ï‡∏≤‡∏£‡∏≤‡∏á

    headers = ["Script", "‡∏Ñ‡∏≥‡πÅ‡∏õ‡∏•‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢", "‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å", "‡πÄ‡∏â‡∏•‡∏¢‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢"]

    # 1. Draw Headers
    cur_x = table_x
    for i, h in enumerate(headers):
        draw.rectangle([cur_x, table_y, cur_x+col_w[i], table_y+header_h], fill=(68, 114, 196), outline=(0,0,0))
        draw_text_box(draw, h, cur_x, table_y+15, col_w[i], header_h, font_head, fill=(255,255,255), align='center')
        cur_x += col_w[i]

    # 2. Prepare Data
    eng_lines = q_data['script_eng'].split('\n')
    thai_lines = q_data['script_thai'].split('\n')
    while len(eng_lines) < 4: eng_lines.append("")
    while len(thai_lines) < 4: thai_lines.append("")

    eng_lines[3] += f"\n(Narrator): {q_data['narrator']}"
    thai_lines[3] += f"\n(Narrator): {q_data.get('narrator_thai','')}"

    # 3. Draw Rows
    start_y = table_y + header_h

    for r in range(4):
        y = start_y + (r * row_h)

        # Col 0: Eng
        draw.rectangle([table_x, y, table_x+col_w[0], y+row_h], outline=(0,0,0))
        draw_text_box(draw, eng_lines[r], table_x, y, col_w[0], row_h, font_content)

        # Col 1: Thai
        x1 = table_x + col_w[0]
        draw.rectangle([x1, y, x1+col_w[1], y+row_h], outline=(0,0,0))
        draw_text_box(draw, thai_lines[r], x1, y, col_w[1], row_h, font_content)

        # Col 2: Choices
        x2 = x1 + col_w[1]
        txt = f"{q_data['choices_eng'][r]}\n{q_data['choices_thai'][r]}" if r < 4 else ""
        draw.rectangle([x2, y, x2+col_w[2], y+row_h], outline=(0,0,0))
        draw_text_box(draw, txt, x2, y, col_w[2], row_h, font_content)

    # Col 3: Answer (Merged)
    x3 = table_x + col_w[0] + col_w[1] + col_w[2]
    total_h = row_h * 4
    draw.rectangle([x3, start_y, x3+col_w[3], start_y+total_h], outline=(0,0,0))
    # ‡∏õ‡∏£‡∏±‡∏ö‡∏™‡∏µ‡πÄ‡∏â‡∏•‡∏¢‡πÉ‡∏´‡πâ‡∏≠‡πà‡∏≤‡∏ô‡∏á‡πà‡∏≤‡∏¢‡∏Ç‡∏∂‡πâ‡∏ô (‡∏™‡∏µ‡πÅ‡∏î‡∏á‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏´‡∏°‡∏π)
    draw_text_box(draw, q_data['answer_text'], x3, start_y, col_w[3], total_h, font_content, fill=(150, 0, 0))

    img.save(filename)
    return filename
```

```python
def create_summary_slide(data_list, filename):
    """‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏†‡∏≤‡∏û Slide ‡πÄ‡∏â‡∏•‡∏¢‡∏ï‡∏≠‡∏ô‡∏ó‡πâ‡∏≤‡∏¢ (‡∏â‡∏ö‡∏±‡∏ö‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç Layout 1920x1080)"""
    W, H = 1920, 1080  # 1. ‡πÅ‡∏Å‡πâ‡∏Ç‡∏ô‡∏≤‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ö Video ‡∏´‡∏•‡∏±‡∏Å
    img = Image.new('RGB', (W, H), color=(20, 20, 20)) # ‡∏û‡∏∑‡πâ‡∏ô‡∏´‡∏•‡∏±‡∏á‡πÄ‡∏ó‡∏≤‡πÄ‡∏Å‡∏∑‡∏≠‡∏ö‡∏î‡∏≥ (‡∏î‡∏π‡∏™‡∏ö‡∏≤‡∏¢‡∏ï‡∏≤‡∏Å‡∏ß‡πà‡∏≤‡∏î‡∏≥‡∏™‡∏ô‡∏¥‡∏ó)
    draw = ImageDraw.Draw(img)

    # --- Title ---
    font_title = get_font(100, bold=True)
    # ‡∏à‡∏±‡∏î‡∏Å‡∏∂‡πà‡∏á‡∏Å‡∏•‡∏≤‡∏á‡πÅ‡∏ô‡∏ß‡∏ô‡∏≠‡∏ô
    title_text = "SUMMARY ANSWERS"

    # ‡∏ß‡∏±‡∏î‡∏Ç‡∏ô‡∏≤‡∏î‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏à‡∏±‡∏î‡∏Å‡∏∂‡πà‡∏á‡∏Å‡∏•‡∏≤‡∏á (‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö PIL ‡πÄ‡∏ß‡∏≠‡∏£‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏Å‡πà‡∏≤/‡πÉ‡∏´‡∏°‡πà)
    if hasattr(font_title, 'getlength'): title_w = font_title.getlength(title_text)
    else: title_w = font_title.getsize(title_text)[0]

    draw.text(((W - title_w) / 2, 80), title_text, fill=(255, 255, 255), font=font_title)

    # --- Grid System ---
    font_text = get_font(60, bold=False) # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡∏ô‡∏≤‡∏î‡∏ü‡∏≠‡∏ô‡∏ï‡πå‡πÉ‡∏´‡πâ‡∏≠‡πà‡∏≤‡∏ô‡∏á‡πà‡∏≤‡∏¢

    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö 1920px)
    col1_x = 500  # ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ã‡πâ‡∏≤‡∏¢
    col2_x = 1200 # ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏Ç‡∏ß‡∏≤
    start_y = 300 # ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÅ‡∏Å‡∏ô Y
    line_height = 120 # ‡∏£‡∏∞‡∏¢‡∏∞‡∏´‡πà‡∏≤‡∏á‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î

    for i, item in enumerate(data_list):
        # 2. ‡πÉ‡∏ä‡πâ Regex ‡∏î‡∏∂‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å (A, B, C, D) ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡πÜ
        match = re.search(r'\(([A-D])\)', item['answer_text'])
        ans_char = match.group(1) if match else "?"

        # ‡∏à‡∏±‡∏î‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏° "Q1 : C"
        text = f"Q{item['id']} :  {ans_char}"

        if i < 5: # 5 ‡∏Ç‡πâ‡∏≠‡πÅ‡∏£‡∏Å (‡∏Ç‡πâ‡∏≠ 1-5)
            # ‡∏ß‡∏≤‡∏î‡∏û‡∏∑‡πâ‡∏ô‡∏´‡∏•‡∏±‡∏á‡πÅ‡∏ñ‡∏ö‡∏™‡∏µ‡∏à‡∏≤‡∏á‡πÜ ‡πÉ‡∏´‡πâ‡∏î‡∏π‡∏á‡πà‡∏≤‡∏¢‡∏Ç‡∏∂‡πâ‡∏ô
            y_pos = start_y + (i * line_height)
            draw.text((col1_x, y_pos), text, fill=(255, 255, 0), font=font_text)
        else: # 5 ‡∏Ç‡πâ‡∏≠‡∏´‡∏•‡∏±‡∏á (‡∏Ç‡πâ‡∏≠ 6-10)
            y_pos = start_y + ((i - 5) * line_height)
            draw.text((col2_x, y_pos), text, fill=(255, 255, 0), font=font_text)

    img.save(filename)
    return filename
```
